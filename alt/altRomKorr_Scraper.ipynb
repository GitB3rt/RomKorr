```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

######################################################################################################################################################
#
#    Der Code extrahiert von den ersten 21000 Briefen die Metadaten und schreibt sie in eine CSV
#
######################################################################################################################################################
```

```python
# Base URL
BASE_URL = "https://briefe-der-romantik.de/letters/view/"

# Function to extract data from a single page
def extract_metadata(letter_id):
    url = f"{BASE_URL}{letter_id}?left=text&query_id=6773ee556fe85"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to fetch page for ID {letter_id}: {response.status_code}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract metadata section
    metadata_section = soup.find("div", id="metadata-tab-pane")
    if not metadata_section:
        print(f"No metadata found for ID {letter_id}")
        return None

    # Extract specific fields
    metadata = {
        "Datum": None,
        "Absender": None,
        "Empfnger": None,
        "Absendeort": None,
        "Empfangsort": None,
        "link": url
    }

    for li in metadata_section.find_all("li"):
        label = li.find("span", class_="label")
        kdata = li.find("span", class_="kdata")

        if not label or not kdata:
            continue

        label_text = label.text.strip()
        kdata_text = kdata.text.split("  ")[0].strip()  # Remove GND/GeoNames links

        if label_text in metadata:
            metadata[label_text] = kdata_text

    return metadata

# Function to iterate over the range of IDs and collect data
def scrape_letters(start_id, end_id):
    letters_data = []

    for letter_id in range(start_id, end_id + 1):
        print(f"Processing letter ID {letter_id}...")
        metadata = extract_metadata(letter_id)

        if metadata:
            letters_data.append(metadata)

        time.sleep(0.5)  # Polite delay to avoid overloading the server

    return letters_data

# Main function
def main():
    # Scrape data for IDs 1 to 21000
    letters_data = scrape_letters(1, 21000)

    # Create a DataFrame from the collected data
    df = pd.DataFrame(letters_data)

    # Save to CSV
    df.to_csv("letters_metadata.csv", index=False)
    print("Data saved to letters_metadata.csv")

    return df

# Main execution
if __name__ == "__main__":
    df = main()

# Display the DataFrame if running interactively
#df.head()
```
